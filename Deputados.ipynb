```
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
import datetime
```

```
URL_2019 = 'https://raw.githubusercontent.com/tcnl/deputados/main/data/Ano-2019.csv'
dep_2019 = pd.read_csv(URL_2019, error_bad_lines=False, sep=';', parse_dates=['datEmissao'])
```

```
dep_2019['txtDescricao'] = dep_2019['txtDescricao'].astype('category')
dep_2019['sgPartido'] = dep_2019['sgPartido'].astype('category')
dep_2019['sgUF'] = dep_2019['sgUF'].astype('category')
dep_2019['txtFornecedor'] = dep_2019['txtFornecedor'].astype('category')
dep_2019['txtCNPJCPF'] = dep_2019['txtCNPJCPF'].astype('category')
dep_2019['txtPassageiro'] = dep_2019['txtPassageiro'].astype('category')
dep_2019['txtTrecho'] = dep_2019['txtTrecho'].astype('category')
dep_2019['txNomeParlamentar'] = dep_2019['txNomeParlamentar'].astype('category')
dep_2019.dropna(how='all', axis='columns', inplace=True)
```
